### Vizualization Techgnique

#### 1. Scatter Plots

````python
import numpy as np
````

````python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000*np.random.rand(100)
````

````python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap=plt.cm.viridis) # colors: of points; sizes: of points; alpha: transparency level
plt.colorbar() # adds a colorbar scale
````

- **```cmap='viridis'```** : Matplotlib provides wide ranges of colormaps.
- An optional Parameter, for creating visually appleaing plots.
- Examples you can set: **```cmap=plt.cm.Blues```**,**```cmap=plt.cm.autumn```**,**```cmap=plt.cm.gist_earth```**

#### 2.Line Plots

````python
year = [1970, 1980, 1990, 2000, 2010, 2020]
pop_A = [4.9, 5.3, 7.1, 10.7, 13.5, 15.6]
pop_B = [44.4, 55.6, 69.7, 87.1, 95.4, 100.5]
plt.style.use('ggplot')
plt.plot(year, pop_A, color='g')
plt.plot(year, pop_B, color='r')
plt.xlabel('Countries')
plt.ylabel('Population in million')
plt.title('Population over the years')

plt.savefig("line.png")
````

#### 3. Histograms

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)
# plt.hist(x=data)
plt.hist(x=data, bins=50, alpha=0.5, histtype='stepfilled', color='steelblue', edgecolor='none')
plt.hist(x=(data+2), bins=60, alpha=0.7, histtype='stepfilled', color='red', edgecolor='none')
plt.savefig("hist1.png")
````

#### 4. Bar Plots

````python
labels = ('Python', 'C++', 'Java', 'Perl', 'C#')
num_users = [12, 8, 15, 4, 6]
index = np.arange(len(labels))

plt.bar(index, num_users, align='center', alpha=0.5)
plt.xticks(index,labels)
plt.xlabel('Language')
plt.ylabel('Num of Users')
plt.title('Programming language usage')

plt.savefig('barplot.png')
````

#### 5. Box Plot

````python
data_group_1 = np.random.normal(100, 10, 200)
data_group_2 = np.random.normal(80, 10, 200)
data_group_3 = np.random.normal(90, 10, 200)

data_to_plot = [data_group_1, data_group_2, data_group_3]

plt.boxplot(data_to_plot, patch_artist= True)

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Some Group Scores')

plt.savefig('boxplot.png')
````
